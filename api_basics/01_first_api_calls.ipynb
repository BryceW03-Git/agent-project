```python
import anthropic

client = anthropic.Anthropic()

response = client.messages.create(
    model="claude-sonnet-4-5",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Say hello in exactly 5 words."}
    ]
)

print(response.content[0].text)
```

```python
response = client.messages.create(
    model="claude-sonnet-4-5",
    max_tokens=1024,
    system="You are a commercial real estate analyst. Be terse and precise. Use industry vocabulary.",
    messages=[
        {"role": "user", "content": "Explain the difference between unlevered and levered IRR in two sentences."}
    ]
)

print(response.content[0].text)
```

```python
conversation = [
    {"role": "user", "content": "I'm analyzing a 200-unit multifamily development in Austin."}
]

response = client.messages.create(
    model="claude-sonnet-4-5",
    max_tokens=1024,
    system="You are a commercial real estate analyst. Ask clarifying questions when useful.",
    messages=conversation
)

# Add Claude's response to the conversation history
conversation.append({"role": "assistant", "content": response.content[0].text})

print(response.content[0].text)
```

```python
print(f"Number of messages in conversation: {len(conversation)}")
print()
for i, msg in enumerate(conversation):
    print(f"[{i}] ROLE: {msg['role']}")
    print(f"    CONTENT: {msg['content'][:100]}...")
    print()
```
